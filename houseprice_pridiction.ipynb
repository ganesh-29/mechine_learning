```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("housePrice.csv")
```

```python
data.head()
```

```python
columns=['Parking','Warehouse','Elevator','Address']
cate_value=data[columns]
coln=['Area','Room']
num_valu=data[coln]
num_valu['Price']=data['Price(USD)']
num_valu
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
cate_value_copy=cate_value.copy()
for col in cate_value_copy.columns:
  cate_value_copy[col]=lb.fit_transform(cate_value_copy[col])
cate_value_copy
```

```python
house_data=pd.concat([cate_value_copy,num_valu],axis=1)
```

```python
house_data.isnull().sum()
```

```python
x=house_data.iloc[:,:-1]
y=house_data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_test,y_test)
lr.fit(x_train,y_train)
p=lr.predict(x_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming lr is your trained Linear Regression model


# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(n, p)
print('Mean Absolute Error:', mae)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(n, p)
print('Mean Squared Error:', mse)

# Calculate R-squared
r2 = r2_score(n, p)
print('R-squared:', r2)
```
